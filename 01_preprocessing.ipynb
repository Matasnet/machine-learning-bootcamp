```python
import numpy as np
import pandas as pd
import sklearn #sckit-learn

sklearn.__version__
```

```python
# Wygenerowanie danych
```

```python
data = {
    'size':['XL','L','M','L','M'],
    'color':['red','green','blue','green','red'],
    'gender':['female','male','male','female','female'],
    'price':[199.0, 89.0, 99.0, 129.0, 79.0],
    'weight':[500, 450, 300, 380, 410],
    'bought':['yes','no','yes','no','yes']
}

df_raw = pd.DataFrame(data=data)
df_raw
```

```python
# Utworznie kopii danych
```

```python
df = df_raw.copy()
df.info()
```

```python
# Zmiana typu danych i wstępna eksploracja
```

```python
for col in ['size','color','gender','bought']:
  df[col] = df[col].astype('category')

df['weight'] = df['weight'].astype('float')

df.info()
```

```python
df.describe() # wyświetla dla liczb
```

```python
df.describe().T
```

```python
df.describe(include=["category"]).T
```

```python
# LabelEncoder - przygotowani danych do uczenia maszynowego
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
le.fit(df['bought'])
le.transform(df['bought'])
```

```python
le.fit_transform(df['bought'])
```

```python
le.classes_ #wyśwpietla mapowanie
```

```python
df['bought'] = le.fit_transform(df['bought']) #przypisanie wartości 0 1 

df
```

```python
#Gdybyśmy chcieli wrócić do poprzedniej wartości bought

```

```python
le.inverse_transform(df['bought'])
```

```python
df['bought'] = le.inverse_transform(df['bought'])
df
```

Standaryzacja - StandardScaler
std() - pandas niebociążony
srd() - numpy obciążony

```python
print(f"{df['price']}\n")
print(f"Średnia: {df['price'].mean()}")
print(f"Odchylenie standardowe: {df['price'].std():.4f}")
```

```python
(df['price'] - df['price'].mean()) / df['price'].std()
```

```python
def standardize(x):
  return (x - x.mean()) / x.std()
  
standardize(df['price'])
```

```python
from sklearn.preprocessing import scale

scale(df['price'])
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(df[['price']])
scaler.transform(df[['price']])
```

```python
scaler = StandardScaler()
df[['price', 'weight']] = scaler.fit_transform(df[['price', 'weight']])
df
```
